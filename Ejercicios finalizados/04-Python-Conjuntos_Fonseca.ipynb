<h3 id="content">Contenido de un Conjunto</h3>

Un conjunto es una colección de objetos únicos en Python. Un conjunto se define usando llaves de apertura y cierre <b>{}</b>. Python removerá automáticamente elementos duplicados.

```python
# Crear un conjunto

set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1
```

El proceso de mapeado se muestra en la siguiente figura:

También se puede crear un conjunto a partir de una lista:

```python
# Convertir una lista en un conjunto

album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)             
album_set
```

Vamos a crear un conjunto de géneros musicales:

```python
# Convertir una lista en un conjunto

music_genres = set(["pop", "pop", "rock", "folk rock", "hard rock", "soul", \
                    "progressive rock", "soft rock", "R&B", "disco"])
music_genres
```

```python
# Convertir una tupla en un conjunto

music_genres = set(("pop", "pop", "rock", "folk rock", "hard rock", "soul", \
                    "progressive rock", "soft rock", "R&B", "disco"))
music_genres
```

```python
music_genres.add((1,3,5))
music_genres
```

```python
music_genres.add([1,3,5])  # Esto genera un error porque los elementos de un conjunto deben ser inmutables
```

### ***Analisis: Contenido de un Conjunto***
Los conjuntos en Python son una colección de datos que pueden incluir distintos tipos de elementos, como cadenas de caracteres, números enteros o números flotantes, e incluso combinarlos entre sí. La condición principal es que los elementos almacenados sean inmutables, es decir, que no puedan modificarse, por eso se puede ver que al intentar agregar listas hay un error, pero con las tuplas no.

Otra característica fundamental de los conjuntos es que no permiten elementos repetidos. Si un mismo dato se intenta agregar más de una vez, el conjunto conservará únicamente una sola copia. En otras palabras, todos los elementos dentro de un conjunto deben ser únicos.

Para definir un conjunto en Python, se utilizan llaves ({ }), a diferencia de las listas, que se crean con corchetes. Esta notación permite identificar claramente que se está trabajando con una estructura de tipo conjunto.

<h3 id="op">Operaciones con Conjuntos</h3>

Vamos a revisar las operaciones con conjuntos que nos ayudaran a modificarlo. Considera el conjunto <b>A</b>:

```python
# Conjunto de ejemplo

A = set(["Thriller", "Back in Black", "AC/DC"])
A
```

Podemos agregar un elemento al conjunto mediante el método <code>add()</code>:

```python
# Agregar un elemento al conjunto

A.add("NSYNC")
A
```

Si agregamos el mismo elemento dos veces nada cambiara debido a que no puede haber elemento duplicados en un conjunto:

```python
# Trata de agregar un elemento duplicado al conjunto

A.add("NSYNC")
A
```

Podemos borrar un elemento del conjunto mediante el método <code>remove</code>:

```python
# Remover un elemento del conjunto

A.remove("NSYNC")
A
```

Podemos verificar si existe un elemento especifico dentro del conjunto con el comando <code>in</code>:

```python
# Verificar si existe un elemento en el conjunto

"AC/DC" in A
```

```python
# fucionar dos conjuntos
N = set([3,8,90,43])
M = set([7,43,90,1000])
N.update(M)
N
```

### ***Analisis: Operaciones con Conjuntos***
En los conjuntos existen varias operaciones principales que permiten realizar distintas acciones sobre sus elementos, como agregar, eliminar o verificar datos.

Para agregar un nuevo elemento a un conjunto se utiliza la operación add(). Para eliminar un elemento, se emplea la operación remove(), la cual también existe en las listas y permite borrar un valor específico del conjunto.

Otra operación importante es in, que se utiliza para verificar si un elemento determinado pertenece o no a un conjunto. Esta operación devuelve un resultado lógico según la presencia del dato evaluado.

Además, existe la función set(), cuyo nombre proviene del término en inglés para conjunto. Esta función permite transformar otras estructuras, como listas, en conjuntos, eliminando automáticamente los elementos repetidos.

Por último, se encuentra la operación update(), que permite fusionar dos conjuntos, ya sea modificando un conjunto existente o incorporando los elementos de un conjunto secundario. Su funcionamiento es similar al de extend() en las listas, ya que agrega múltiples elementos en una sola operación.

<h3 id="logic">Operacion Lógicas con Conjuntos</h3>

Recuerda que puedes revisar la diferencia entre dos conjuntos, así como la diferencia simétrica, intersección y unión:

Considera estos dos conjuntos:

```python
# Conjuntos de ejemplo

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

```python
# Imprimir dos conjuntos

album_set1, album_set2
```

Como los dos conjuntos contienen <b>AC/DC</b> y <b>Back in Black</b> podemos representar estos elementos en común en la intersección de dos círculos.

Para encontrar la intersección de dos conjuntos usaremos <code>&</code>:

```python
# Encontrar las intersecciones

intersection = album_set1 & album_set2
intersection
```

Puedes encontrar los elementos que únicamente están dentro de <code>album_set1</code> usando el método <code>difference</code>:

```python
# Encuentra la diferencia en set1 pero no en set2

album_set1.difference(album_set2)  
```

Solo tienes que considerar los elementos en <code>album_set1</code>; todos los de <code>album_set2</code>, incluyendo los de la intersección, no están incluidos.

Puedes encontrar la intersección de <code>album_list1</code> y <code>album_list2</code> con el método <code>intersection</code>:

```python
# Usar el método intersection para encontrar la intersección entre of album_list1 y album_list2

album_set1.intersection(album_set2)   
```

Lo siguiente corresponde a la intersección de los dos círculos:

La unión es la suma de todos los elementos de ambos conjuntos, se representa pintando los dos círculos:

La unión esta dada por:

```python
# Encontrar la unión de los dos conjuntos

album_set1.union(album_set2)
```

Se puede averiguar si un conjunto es un súper conjunto o un sub conjunto de otro, respectivamente, de la siguiente forma:

```python
# Revisar si es un super conjunto

set(album_set1).issuperset(album_set2)   
```

```python
# Revisar si es un sub conjunto

set(album_set2).issubset(album_set1)     
```

Este es un ejemplo donde <code>issubset()</code> y <code>issuperset()</code> regresan un valor verdadero:

```python
# Revisar si es un sub conjunto

set({"Back in Black", "AC/DC"}).issubset(album_set1) 
```

```python
# Revisar si es un super conjunto

album_set1.issuperset({"Back in Black", "AC/DC"})   
```

### ***Analisis: Operacion Lógicas con Conjuntos***
Tal como se ha visto, los conjuntos en Python cumplen la misma función que los conjuntos matemáticos trabajados de forma teórica. Permiten realizar operaciones lógicas y matemáticas como la intersección, la unión y otras relaciones propias de los conjuntos.

Para realizar la intersección entre dos conjuntos existen dos opciones: utilizar el símbolo amperstand (&) o el método intersection(). Ambas permiten obtener únicamente los elementos que se encuentran en común entre los conjuntos.

La unión se realiza mediante el método union(), el cual devuelve un conjunto que contiene todos los elementos presentes en ambos conjuntos, sin repetir valores.

Otra operación importante es difference(), que permite obtener los elementos que pertenecen exclusivamente a un conjunto, excluyendo aquellos que forman parte de la intersección con otro conjunto.

Por último, mediante los métodos issuperset() e issubset() es posible verificar relaciones entre conjuntos, determinando si un conjunto es un superconjunto (contiene a otro conjunto) o un subconjunto (está contenido dentro de otro).

Gracias a estas operaciones, los conjuntos permiten realizar distintos tipos de análisis lógico-matemático de forma clara y eficiente dentro de Python.

### ***Conclusión***
En conclusión, aunque los conjuntos presenten ciertas similitudes con las listas y las tuplas en el sentido de que pueden contener datos, sus funciones y objetivos son drásticamente diferentes.

Los conjuntos están diseñados específicamente para almacenar datos únicos, es decir, datos que no se repiten, y utilizar estos valores para realizar análisis y comparaciones con otros conjuntos. A partir de este enfoque se pueden llevar a cabo operaciones como la unión, la intersección y la diferencia, con el fin de obtener conclusiones sobre conjuntos de datos.

Debido a esta finalidad, los conjuntos no permiten datos mutables. Por esta razón, las listas no pueden formar parte de un conjunto. En cambio, las tuplas, al ser estructuras inmutables, sí pueden almacenarse dentro de los conjuntos y participar en este tipo de análisis.

Por lo tanto, se puede afirmar que los conjuntos son estructuras orientadas principalmente a la lógica y las operaciones matemáticas, enfocadas en la comparación y el análisis de datos únicos, más que en el almacenamiento tradicional de información.
