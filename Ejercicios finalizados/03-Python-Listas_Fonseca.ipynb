<h2 id="list">Listas</h2>

<h3 id="index">Indexación</h3>

Vamos a ver un poco mas sobre las listas en Python. Una lista es una colección ordenada de distintos objetos tales como enteros, cadenas e incluso otras listas. La dirección de cada elemento dentro de una lista se llama <b>índice</b>. El índice se usa para acceder o referirse a un objeto dentro de la lista.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsIndex.png" width="1000" />

Para crear una lista, se escribe cada elemento dentro de <b>[ ]</b>, separado cada uno por comas. ¡Vamos a probar!

```python
# Crear una lista

L = ["Michael Jackson", 10.1, 1982]
L
```

Podemos usar en una lista la indexación regular o negativa:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsNeg.png" width="1000" />

```python
# Imprime los elementos de cada índice

print('the same element using negative and positive indexing:\n Postive:',L[0],
'\n Negative:' , L[-3]  )
print('the same element using negative and positive indexing:\n Postive:',L[1],
'\n Negative:' , L[-2]  )
print('the same element using negative and positive indexing:\n Postive:',L[2],
'\n Negative:' , L[-1]  )
```

### ***Analisis: Indexación***
La indexación en listas presenta la misma funcionalidad que la vista anteriormente en las tuplas. Aunque pueda resultar repetitivo, el principio se mantiene: los índices se crean y utilizan de la misma manera.

En las listas existe la indexación normal, que va desde 0 hasta n - 1, y la indexación negativa, que va desde -n hasta -1, donde n representa la cantidad total de elementos contenidos en la lista. El acceso a los elementos se realiza de forma idéntica a la de las tuplas.

Una diferencia importante entre ambos tipos de estructuras es su forma de definición. Mientras que las tuplas se determinan mediante paréntesis (( )), las listas se generan utilizando corchetes ([ ]).

<h3 id="content">Contenido de la Lista</h3>

Las listas pueden contener cadenas, números enteros y en punto flotante. Podemos anidar otras listas y además tuplas u otros tipos de estructuras de datos. Para el anidado se siguen las mismas reglas:

```python
# Lista de ejemplo

["Michael Jackson", 10.1, 1982, [1, 2], ("A", 1)]
```

```python
# Se pueden buscar atyravez del indice dentro de tuplas o listas, etc que esten dentro de una lista mas grando.
Things = [10, ("Arroz", 66), ['GTA', 'Libro']]
Things[2][1]
```

### ***Analisis: Contenido de la Lista***
Las listas pueden contener diferentes tipos de datos, de forma similar a las tuplas. Dentro de una lista es posible almacenar cadenas de caracteres, números enteros, números flotantes, otras listas y también tuplas.

En cuanto al contenido, las listas y las tuplas funcionan de manera muy parecida, ya que ambas permiten agrupar datos de distintos tipos dentro de una misma estructura. No existe una restricción que impida a una tupla contener una lista, ni a una lista contener una tupla.

Más adelante se pueden identificar con mayor claridad las verdaderas diferencias entre listas y tuplas, las cuales no están relacionadas con el tipo de datos que pueden almacenar, sino con la forma en que dichos datos pueden modificarse.

<h3 id="op">Operaciones con Listas</h3>

También podemos aplicar el slicing en las listas. Por ejemplo, si queremos obtener los últimos dos elementos, usamos el siguiente comando:

```python
# Lista de ejemplo

L = ["Michael Jackson", 10.1,1982,"MJ",1]
L
```

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsSlice.png" width="1000">

```python
# Slicing en una lista

L[3:5]
```

Podemos usar el método <code>extend</code> para añadir elementos a la lista:

```python
# Aplica el método extend para añadir elementos

L = [ "Michael Jackson", 10.2]
L.extend(['pop', 10])
L
```

Otro método similar es <code>append</code>. Si usamos <code>append</code> en vez de <code>extend</code>, añadimos un elemento a a la lista:

```python
# Aplica el método append para añadir elementos

L = [ "Michael Jackson", 10.2]
L.append(['pop', 10])
L
```

Cada vez que un método es aplicado, la lista cambia. Al usar <code>extend</code> estamos añadiendo dos nuevos elementos. La lista <code>L</code> es entonces modificada con dos nuevos elementos.

```python
# Aplica el método extend para añadir elementos

L = [ "Michael Jackson", 10.2]
L.extend(['pop', 10])
L
```

Si añadimos a la lista <code>['a','b']</code> tendremos un nuevo elemento conformado por una lista anidada:

```python
# Aplica el método append para añadir elementos

L.append(['a','b'])
L
```

Las listas son estructuras que pueden cambiar con el tiempo. Por ejemplo, cambiemos el primer elemento de una lista de la siguiente manera:

```python
# Cambia el elemento usando el índice

A = ["disco", 10, 1.2]
print('Before change:', A)
A[0] = 'hard rock'
print('After change:', A)
```

Para borrar un elemento de la lista, usamos el comando <code>del</code>:

```python
# Borra el elemento usando el índice

print('Before change:', A)
del(A[0])
print('After change:', A)
```

Podemos transformar una cadena en una lista con <code>split</code>. Por ejemplo, el método <code>split</code> convierte cada grupo de caracteres separados por un espacio en blanco, en un elemento de la lista:

```python
# Divide la cadena

'hard rock'.split()
```

Podemos utilizar la función split para separar distintas cadenas usando algún carácter especifico. Pasamos en el argumento el carácter que queremos usar como separador, en este caso será una coma. El resultado es una lista, y cada elemento corresponderá a un grupo de caracteres separados por una coma:

```python
# Separa la cadena usando comas

'A,B,C,D'.split(',')
```

```python
lista = [10, 20, 30, 20]
lista.remove(20)
print(lista)
```

```python
x = lista.pop(1)
print(x)     
print(lista)
```

```python
lista = [4, 1, 3, 2]
lista.sort()
print(lista)
```

### ***Analisis: Operaciones con Listas***
Al hablar de operaciones con listas, se puede identificar que, a diferencia de las tuplas, las listas sí pueden modificarse de diversas maneras. Esta característica las convierte en una estructura de datos más flexible.

Las listas pueden extenderse mediante el método .extend(), el cual permite fusionar dos listas en una sola. También pueden ampliarse utilizando .append(), que sirve para introducir un nuevo elemento dentro de una lista, incluyendo otra lista como un solo elemento.

Además, existe el comando del, que permite eliminar elementos de una lista utilizando el índice proporcionado por la indexación. Este mismo mecanismo de indexación también posibilita modificar valores directamente dentro de la lista.

Otro comando importante es split(), el cual permite transformar una cadena de caracteres en una lista. Este proceso puede realizarse de forma predeterminada, separando los elementos por espacios en blanco, o especificando qué carácter o conjunto de caracteres se utilizará como separador.

Fuera de los comandos presentados inicialmente, se identificaron otros métodos relevantes:

remove(), que elimina un elemento de la lista indicando directamente su valor. Si el valor aparece más de una vez, solo se eliminará la primera coincidencia encontrada.

pop(), que elimina un elemento de la lista, pero a diferencia de otros métodos, devuelve el valor eliminado, permitiendo almacenarlo en otra variable.

sort(), que permite ordenar los elementos de una lista, siempre y cuando todos los datos sean del mismo tipo y puedan compararse entre sí.

En conjunto, estas operaciones evidencian la versatilidad y potencia de las listas para la manipulación y organización de datos en Python.

<h3 id="co">Copiado y Clonación de Listas</h3>

Cuando asignamos el valor de la variable <b>B</b> igual a <b>A</b>; ambas están haciendo referencia a la misma lista guardada en memoria:

```python
# Copiar (por referencia) la lista A

A = ["hard rock", 10, 1.2]
B = A
print('A:', A)
print('B:', B)
```

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsRef.png" width="1000" align="center">

En un principio, el valor del primer elemento en <b>B</b> es asignado como hard rock. Si modificamos el primer elemento en <b>A</b> por <b>banana</b>, obtendremos un efecto inesperado. Debido a que <b>A</b> y <b>B</b> hacen referencia a la mima lista, si cambiamos <b>A</b>, entonces cambiamos también a <b>B</b>. Al revisar el primer elemento de <b>B</b> veremos banana en vez de hard rock:

```python
# Examinar la copia hecha por referencia

print('B[0]:', B[0])
A[0] = "banana"
print('B[0]:', B[0])
```

Se muestra esto en la siguiente imagen:

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsRefGif.gif" width="1000" />

Se puede clonar la lista **A** haciendo uso de la siguiente sintaxis:

```python
# Clonar la lista A

B = A[:]
B
```

La variable **B** es una copia o clon por valor de la lista original; esto se puede demostrar en la siguiente imagen:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsVal.gif" width="1000" />

Ahora cualquier cambio en <b>A<b/>, no será aplicado a <b>B</b>:

```python
print('B[0]:', B[0])
A[0] = "hard rock"
print('B[0]:', B[0])
```

### ***Analisis: Copiado y Clonación de Listas***
Una diferencia importante entre listas y tuplas se presenta al momento de asignarlas a una nueva variable. En el caso de las listas, cuando una variable se iguala a otra que ya contiene una lista, no se crea una lista nueva. En su lugar, ambas variables pasan a apuntar a la misma lista en memoria.

Esto significa que, si una variable B se iguala a otra variable A y A contiene una lista, cualquier modificación realizada sobre A también afectará a B, ya que ambas referencias apuntan al mismo objeto.

Para clonar una lista y crear una nueva lista independiente con el mismo contenido, es necesario igualar la nueva variable a un slicing completo de la lista original. De esta forma, se copia todo el contenido sin conservar la referencia a la lista original, evitando que los cambios se reflejen en ambas variables.

En contraste con esto, aunque las tuplas también pueden ser asignadas a otra variable, no presentan este problema práctico, ya que son estructuras inmutables y no pueden modificarse una vez creadas. Por esta razón, no es necesario clonarlas para evitar cambios compartidos.

### ***Conclusión***
Para finalizar, se puede observar que las listas presentan, a primera vista, una funcionalidad similar a la de las tuplas. Ambas estructuras pueden contener cualquier tipo de dato, incluyendo otras listas o tuplas, utilizan un sistema de indexación equivalente y permiten operaciones como el slicing.

Sin embargo, la diferencia fundamental entre ambas estructuras radica en su mutabilidad. Las listas pueden modificarse de múltiples formas: es posible cambiar elementos, eliminarlos, agregarlos o extraerlos. En contraste, las tuplas no permiten la modificación de los datos que contienen. Aunque es posible crear nuevas tuplas o unir varias de ellas, no se pueden alterar directamente los elementos de una tupla ya existente sin recrearla por completo.

Debido a esta característica, se puede afirmar que las tuplas son estructuras inmutables, mientras que las listas son estructuras mutables. Este comportamiento implica que las listas suelen ocupar más espacio en memoria y requieren una mayor gestión interna, pero a cambio ofrecen una mayor versatilidad, siendo ideales para trabajar con datos que cambian constantemente y necesitan ser reorganizados o actualizados.

Por otro lado, las tuplas resultan más adecuadas para almacenar datos estáticos, que no deben modificarse una vez definidos, ya que son más ligeras y eficientes en términos de almacenamiento. En consecuencia, cada estructura cumple un propósito específico dependiendo del tipo de datos y del nivel de flexibilidad que se requiera.
